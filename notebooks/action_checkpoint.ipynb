# Director-Class AI Action Checkpoint

This notebook demonstrates the dry-run action-control path. It reviews one safe command and one destructive command, then inspects redacted route metadata. The output is local functional evidence, not a comparative benchmark claim.

```python
from director_class_ai.cli.guard import CommandGuardOptions, run_guard

safe = run_guard(CommandGuardOptions(surface="shell", command=("printf", "ok")))
blocked = run_guard(CommandGuardOptions(surface="shell", command=("rm", "-rf", "/")))

(safe["route"], safe["executed"], blocked["route"], blocked["executed"], blocked["firing"])
```

The safe command should route to `allow` without execution because the guard defaults to dry-run. The destructive command should not execute and should include a destructive-command firing.

```python
assert safe["route"] == "allow"
assert safe["executed"] is False
assert blocked["executed"] is False
assert "destructive_command" in blocked["firing"]

{"safe": safe, "blocked": blocked}
```
